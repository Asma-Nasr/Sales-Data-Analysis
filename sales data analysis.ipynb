### Import pandas

```python
import pandas as pd
```

### Load the dataset

```python
data = pd.read_csv("sales_data.csv")
```

```python
data.head()
```

### Questions

1. What was the total revenue generated by the company over the course of the year?


```python
# 1. Total Revenue
total_revenue = data['revenue'].sum()
print(f"Total Revenue: ${total_revenue:.2f}")
```

2. Which product had the highest revenue? How much revenue did it generate?

```python
# 2. Product with Highest Revenue
highest_revenue_product = data.loc[data['revenue'].idxmax()]
highest_revenue_product_name = highest_revenue_product['product']
highest_revenue_value = highest_revenue_product['revenue']
print(f"Highest Revenue Product: {highest_revenue_product_name} (${highest_revenue_value:.2f})")
```

3. What was the average price of a product sold by the company?

```python
# 3. Average Price
average_price = data['price'].mean()
print(f"Average Price: ${average_price:.2f}")
```

4. What was the total quantity of products sold by the company?

```python
# 4. Total Quantity Sold
total_quantity_sold = data['quantity'].sum()
print(f"Total Quantity Sold: {total_quantity_sold}")
```

5. Which category had the highest revenue? How much revenue did it generate?

```python
# 5. Category with Highest Revenue
category_revenue = data.groupby('category')['revenue'].sum()
highest_revenue_category = category_revenue.idxmax()
highest_revenue_category_value = category_revenue.max()
print(f"Highest Revenue Category: {highest_revenue_category} (${highest_revenue_category_value:.2f})")
```

6. What was the average revenue per sale?

```python
# 6. Average Revenue Per Sale
average_revenue_per_sale = data['revenue'].mean()
print(f"Average Revenue Per Sale: ${average_revenue_per_sale:.2f}")
```

7. What was the total revenue generated in each quarter of the year? (i.e. Q1, Q2, Q3, Q4)

```python
# 7. Total Revenue by Quarter
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)
quarterly_revenue = data.resample('Q')['revenue'].sum()
print("Total Revenue by Quarter:")
print(quarterly_revenue)
```
